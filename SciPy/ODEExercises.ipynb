### Question 1

(a) See what happens to the period of the pendulum as you change the initial angle of release. Plot θ vs t for 5 different initial angles, θ0 , spanning 0 to π. (i.e. 0, π/4 , π/2 , 3π/4 , and π ). Put them all on the same plot. What happens to the period of the pendulum as the initial angle increases from π/4 to 3π/4? What happens to the trajectory when the initial angle is 0 and π ?

```python
def deriv(theta_current,t):  # return derivatives of the array theta
    g = 9.8 #m/s^2
    ell = 0.5 #m
    theta = theta_current[0]
    omega = theta_current[1]
    dtheta = omega
    domega = -(g/ell)*np.sin(theta)
    return np.array([ dtheta, domega ])
```

```python
t = np.arange(0.,10.+0.01,0.01)
omega0 = 0. #starting from rest, omega = 0

theta = []
th0 = arange(0,np.pi+np.pi/4,np.pi/4)

theta_init = np.array([th0[0], omega0]) # initial values
theta = odeint(deriv,theta_init,t) 

theta_init = np.array([th0[1], omega0]) # initial values
theta1 = odeint(deriv,theta_init,t)

theta_init = np.array([th0[2], omega0]) # initial values
theta2 = odeint(deriv,theta_init,t)

theta_init = np.array([th0[3], omega0]) # initial values
theta3 = odeint(deriv,theta_init,t)

theta_init = np.array([th0[4], omega0]) # initial values
theta4 = odeint(deriv,theta_init,t)

plt.plot(t,theta[:,0],'r',label="theta0 = 0")
plt.plot(t,theta1[:,0],'b',label="theta0 = pi/4")
plt.plot(t,theta2[:,0],'g',label="theta0 = pi/2")
plt.plot(t,theta3[:,0],'k',label="theta0 = 3pi/4")
plt.plot(t,theta4[:,0],'m',label="theta0 = pi")
plt.xlabel(r'$t (seconds)$',fontsize=20)
plt.ylabel(r'$\theta (rad)$',fontsize=20)
plt.grid()
plt.legend()
plt.show()
```

As the initial angle increases from pi/4 to 3pi/4, the period increases. When the inital angle is 0 and pi, the pendulum does not move.

(c) For the θ0 = 0.95π case, check that energy is conserved by plotting E/m vs. t (where E/m=gℓ(1−cos(θ))+12ℓ2ω2 ).
Is the energy a constant? If not, then the computational errors are producing a non-physical loss (or gain) of energy. Review the documentation for odeint and change the error tolerance value to a tighter number.

```python
omega = theta[:,1]
Theta = theta[:,0]
g = 9.8
L = 0.5
func = lambda omega,Theta : g*L*(1-np.cos(Theta) + (1/2)*(L**2)*(omega**2))
Edm = func(omega,Theta) 
plt.plot(t,Edm);
plt.xlabel(r'$t (seconds)$',fontsize=20);
plt.ylabel(r'$E/m $',fontsize=20);
```
