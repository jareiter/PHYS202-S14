# Battleship Assignment

Run this code by pressing Shift+Return on the cell beneath this. You will be asked for the number of guesses you would like to make. Then provide the 'x' and 'y' coordinate for each of your guesses.

Your guesses appear in yellow, your hit in red, and the actual ship in black.

For your reference, the Battleship can appear in any orientation on the map (including diagonally) and can be between 1 and 4 blocks long.

```python
from random import randint
from ipythonblocks import BlockGrid
from numpy import size
from numpy import ones
from sys import exit

grid = BlockGrid(30, 30, fill=(0, 255, 255))
grid.show()

print "Let's play Battleship!"

l = randint(0,3)
orientation = randint(0,2)
def random_row(grid):
    return randint(0, grid.height - 4)

def random_col(grid):
    return randint(0, grid.width - 4)

a = random_row(grid)
b = random_col(grid)

if orientation == 0:
    ship_row = range(a,a+randint(1,4))
    ship_col = range(b,b+randint(1,4))
    c = range(size(ship_row))

elif orientation == 1:
    ship_row = range(a,a+randint(1,4))
    ship_col = b
    c = range(size(ship_row))

else:
    ship_col = range(b,b+randint(1,4))
    ship_row = a
    c = range(size(ship_col))

turn_total = int(raw_input("How many guesses would you like? "))
turn = 0
for turn in range(turn_total):
    print "Turn", turn + 1
    guess_row = int(raw_input("Guess Row:"))
    guess_col = int(raw_input("Guess Col:"))
    
    if orientation == 0:
        for row in ship_row:
            for col in ship_col:
                if guess_row == row and guess_col == col:
                    print "Congratulations! You sunk my battleship!"
                    for i in c:
                        grid[ship_row[i],ship_col[i]] = (0,0,0)
                    grid[guess_row,guess_col] = (255,0,0)
                    grid.show()
                    exit(0)

        if (guess_row < 0 or guess_row > 29) or (guess_col < 0 or guess_col > 29):
            print "Oops, that's not even in the ocean."
        elif(grid[guess_row,guess_col] == (225,225,0)):
            print "You guessed that one already."          
        else:
            print "You missed my battleship!"
            grid[guess_row,guess_col] = (255,255,0)
            grid.show()
            if turn == turn_total-1:
                print "Game Over. Check out the location of the Battleship"
                for i in c:
                    grid[ship_row[i],ship_col[i]] = (0,0,0)
                grid.show()
                                
    elif orientation == 1:
        for row in ship_row:
            if guess_row == row and guess_col == ship_col:
                print "Congratulations! You sunk my battleship!"
                for i in c:
                    grid[ship_row[i],ship_col] = (0,0,0)
                grid[guess_row,guess_col] = (255,0,0)
                grid.show()
                exit(0)

        if (guess_row < 0 or guess_row > 29) or (guess_col < 0 or guess_col > 29):
            print "Oops, that's not even in the ocean."
        elif(grid[guess_row,guess_col] == (225,225,0)):
            print "You guessed that one already."          
        else:
            print "You missed my battleship!"
            grid[guess_row,guess_col] = (255,255,0)
            grid.show()
            if turn == turn_total-1:
                print "Game Over. Check out the location of the Battleship"
                for i in c:
                    grid[ship_row[i],ship_col] = (0,0,0)
                grid.show()
                        
    elif orientation == 2:
        for col in ship_col:
            if guess_row == ship_row and guess_col == col:
                print "Congratulations! You sunk my battleship!"
                for i in c:
                    grid[ship_row,ship_col[i]] = (0,0,0)
                grid[guess_row,guess_col] = (255,0,0)
                grid.show()
                exit(0)

        if (guess_row < 0 or guess_row > 29) or (guess_col < 0 or guess_col > 29):
            print "Oops, that's not even in the ocean."
        elif(grid[guess_row,guess_col] == (225,225,0)):
            print "You guessed that one already."          
        else:
            print "You missed my battleship!"
            grid[guess_row,guess_col] = (255,255,0)
            grid.show()
            if turn == turn_total-1:
                print "Game Over. Check out the location of the Battleship"
                for i in c:
                    grid[ship_row,ship_col[i]] = (0,0,0)
                grid.show()
                        
                
                            
        
        
```
