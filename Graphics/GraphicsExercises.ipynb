```python
"""
Demo of scatter plot on a polar axis.

Size increases radially in this example and color increases with angle (just to
verify the symbols are being scattered correctly).
"""
import numpy as np
import matplotlib.pyplot as plt


N = 150
r = 2 * np.random.rand(N)
theta = 2 * np.pi * np.random.rand(N)
area = 200 * r**2 * np.random.rand(N)
colors = theta

ax = plt.subplot(111, polar=True)
c = plt.scatter(theta, r, c=colors, s=area, cmap=plt.cm.hsv)
c.set_alpha(0.75)

plt.show()
```

5.Determine the electric potential for a quadrupole consisting of
q1 = -1.0 nC at (-2 m, 0 m)
q2 = -1.0 nC at (+2 m, 0 m)
q3 = +1.0 nC at (0 m, -2 m)
q4 = +1.0 nC at (0 m, +2 m)
(a) Make a surface plot on the range -4 <(x,y)< 4 using 101 bins in x and 101 bins in y. In this case, do not let k = q = d = 1. Use dimensionful quantities and compute the value of the potential in volts.
Recall k = 8.9875517873681764 × 109 N⋅ m2/C2.

```python
dx = 0.08
dy = 0.08
x = np.arange(-4,4,dx)
y = np.arange(-4,4,dy)
x,y = np.meshgrid(x,y)

k =  8.9875517873681764*10**9; q = 1*10**-9; d = 2;
V=(k*q/(x**2+(y-d)**2)**(1/2.)) - (k*q/(x**2+(y+d)**2)**(1/2.))

fig = figure()

ax = fig.gca(projection='3d')
p = ax.plot_surface(x,y,V,rstride=1,cstride=1,linewidth=0,cmap=cm.coolwarm)

cb = fig.colorbar(p, shrink=0.5)

xlabel("x")
ylabel("y")
```

(b) Now make a contour plot of the same data, with appropriate contour levels to see the shape of the potential.

```python
levels = array([8, 4, 2, 1, 0.5, 0.25, 0, -0.25, -0.5, -1, -2, -4, -8])
plt.contour(x,y,V,levels,linewidths=4,cmap=cm.coolwarm);
```
